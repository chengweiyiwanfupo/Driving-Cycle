```python
#提取一个轨迹文件的运动学片段的特征值
import pandas as pd
import numpy as np
import xlsxwriter as xw

result = pd.read_csv("D:\jupyter notebook 保存文件\测试运动学片段.csv")
result = pd.DataFrame(result)


data = pd.read_csv("D:\西安22运行数据\运行 22\预处理\C001951294036.csv")
data = pd.DataFrame(data)


print("开始计算特征值")
feature = []
#计算特征值
for i in range(len(result)):
    
    m = result.iloc[i, 0]  # 提取运动学片段开始点
    
    n = result.iloc[i, 1]  # 提取运动学片段怠速状态结束点
    
    k = result.iloc[i, 2]+1  # 提取运动学片段结束点+1，因为后面要用到减法才+1
    
    v_average = sum(data.loc[m:k-1,'车速'])/(k-m)#整个运动学片段内的平均速度
    
    v_average_travel = sum(data.loc[n:k-1,'车速'])/(k-n)#运动学片段内的运动段的平均速度
    
    v_max = max(data.loc[m:k-1,'车速'])#整个运动学片段内的最大速度
    
    #加减速
    plus = []
    slow = []
    uniformity = []
    for j in range(m,k):
        acc = data.loc[j, '加速度(m/s^2)']
        v = data.loc[j, '车速']
        if acc >=0.15 and v!=0:
            plus.append(acc)
        if acc <=-0.15 and v!=0:
            slow.append(acc)
        if acc <= 0.15 and acc >= -0.15 and v!=0:
            uniformity.append(acc)
            
    # 平均加速度
    #plus_average = sum(plus)/len(plus)
    # 平均减速度
    #slow_average = sum(slow)/len(slow)
    # 平均加速度
    
    if len(plus)==0:
        plus_average =0
        plus_std =0
    else:
        plus_average = sum(plus)/len(plus)
        plus_std = np.std(plus)
    # 平均减速度
    if len(slow)==0:
        slow_average =0
        slow_std = 0
    else:
        slow_average = sum(slow)/len(slow)
        slow_std = np.std(slow)
    
    # 速度标准差
    v_std = np.std(data.loc[m:k-1,'车速'])
    
    # 加速度标准差
    #plus_std = np.std(plus)
    
    # 减速度标准差
    #slow_std = np.std(slow)
    
    # 运动学片段时长
    time = k - m
    
    # 怠速(idling)时间
    time_i = n - m+1
    
    # 加速时间
    time_p = len(plus)
    
    # 减速时间
    time_s = len(slow)
    
    # 匀速时间
    time_u = len(uniformity)
    
    # 怠速时间比
    ratio_i = time_i/time
    
    # 加速时间比
    ratio_p = time_p / time
    
    # 减速时间比
    ratio_s = time_s / time
    
    # 匀速时间比
    ratio_u = time_u / time
    
    all = [v_average,v_average_travel,v_max,plus_average,slow_average,v_std,plus_std,slow_std,time,time_i,time_p,time_s,time_u,ratio_i,ratio_p,ratio_s,ratio_u]
    feature.append(all)    
    
# 指定列名
columns = ['v_average','v_average_travel','v_max','plus_average','slow_average','v_std','plus_std','slow_std','time','time_i','time_p','time_s','time_u','ratio_i','ratio_p','ratio_s','ratio_u']

# 将列表转换为DataFrame
feature = pd.DataFrame(feature, columns=columns)

# 现在你可以对DataFrame进行操作，并将其保存为CSV或其他格式
feature.to_csv("测试特征值.csv", index=False)
feature
```
