```python
#需要得到每秒的瞬时油耗,对于一个短行程，第一行和最后一行的瞬时油耗为0，中间每一行的瞬时油耗为下一行减这一行
import pandas as pd
df=df_z.copy()
#原本的OBD数据的油量戳单位是10ml，
'''df.loc[:145, '油量戳'] -= 331
df.loc[146:332, '油量戳'] -= 16148
df.loc[333:460, '油量戳'] -= 25566
df.loc[461:547, '油量戳'] -= 16611'''
df['油量戳'] =df['油量戳'] /10


# 创建新的一列瞬时油耗

#瞬时油耗的单位为ml/s
df['瞬时油耗ml/s'] = 1
i=1
while i<len(df)-1:
    
    if (df.at[i,"子行程id"]==df.at[i+1,"子行程id"])and (df.at[i,"行程id"]==df.at[i+1,"行程id"]):
        df.at[i,'瞬时油耗ml/s']=df.at[i+1,'油量戳']-df.at[i,'油量戳']
        
    
    else :
        df.loc[i, '瞬时油耗ml/s'] = 0
        df.loc[i+1,'瞬时油耗ml/s'] =0
        i=i+1
    i+=1    
df.loc[0,'瞬时油耗ml/s'] = 0
df.loc[len(df)-1, '瞬时油耗ml/s'] = 0

df
```
